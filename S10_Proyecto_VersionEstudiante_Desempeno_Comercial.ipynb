#### Vista Overview - ¿Cómo ha evolucionado el ingreso total entre 2024 y 2025?

Especifica qué KPIs incluirás y por qué: (3-4 KPIs):


- **Ventas**: Métrica que responde a la pregunta. Podemos ver la cantidad de ingresos generados en el período evaluado.
- **Utilidad Bruta**: Permite conocer cuánto dinero queda tras los costos. Se obtiene al calcular la diferencia que hay entre `Ingresos` y `Costo`.
- **Margen Bruto**: Dato que evalua si la operación es eficiente, no solo grande. Es la ganancia expresada como porcentaje del ingreso. Se calcula Margen Bruto % = (`Ingresos` − `Costo`)/`Ingresos` * 100.  
- **Unidades Vendidas**: Para separar el efecto precio del efecto volumen. Si el ingreso sube pero las unidades bajan, el crecimiento viene del precio, no de más clientes: eso es una señal estratégica importante.

Explica qué gráfico usarás para cada pregunta y justifica la elección.

#### 🖥️ Vista General (Overview)

**S (Situación):**  
El negocio cerró el período 2024–2025 con ingresos totales de 5,5M, una utilidad bruta de 1,9M y un margen del 35,1% — cifras que reflejan una operación rentable y estable en sus tres países y cuatro categorías de producto.

**C (Complicación):**  
Sin embargo, ese resultado global oculta una vulnerabilidad estructural: el ingreso no está distribuido de forma equilibrada a lo largo del año. Enero y diciembre concentran los picos, el segmento Premium sostiene la mayor parte del ingreso, y Perú lidera con una brecha significativa sobre Chile y Colombia. El negocio depende de pocos meses, pocos clientes y un solo país para sostener sus números.

**Q (Pregunta):**   
¿Es sostenible un modelo comercial donde una sola estación, un solo segmento y un solo país explican la mayor parte del resultado?

**A (Respuesta):**  
No en el largo plazo. El dashboard revela que el patrón se repite igual en 2024 y 2025 — lo que confirma que no es un evento puntual sino la estructura actual del negocio. La rentabilidad es real, pero frágil. Una mala temporada de verano o una pérdida de clientes Premium impacta directamente el resultado anual sin mecanismos visibles de compensación.

#### 🔎 Vista Detalle

**S (Situación):**  
Al profundizar en los datos, la caída de ingresos en mitad de año es consistente en todos los países y todas las categorías. Entre junio y agosto el negocio opera en su punto más bajo, con ingresos que representan menos de la mitad del pico de enero.

**C (Complicación):**  
El precio unitario no es el problema — se mantiene estable durante todo el año. Lo que colapsa es el volumen de unidades vendidas. Esto descarta una guerra de precios como causa y confirma que el negocio enfrenta una caída de demanda estacional que hasta ahora no tiene respuesta comercial. Además, la tabla de detalle muestra que ningún país ni categoría escapa a este patrón: Chile, Colombia y Perú comparten la misma estructura de resultados, sin un caso atípico que distorsione el conjunto.

**Q (Pregunta):**  
¿Qué palancas concretas — por país, categoría o segmento — pueden activar la demanda en los meses débiles sin sacrificar el margen que hoy sostiene el negocio?

Segundo, las categorías tienen rentabilidades distintas que aún no se están usando como criterio de priorización comercial. Tercero, el segmento Económico está subdesarrollado y podría activarse en períodos de baja demanda sin canibalizar al segmento Premium.
